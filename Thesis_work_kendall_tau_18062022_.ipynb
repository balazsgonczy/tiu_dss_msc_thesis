```python
#Load data from dropbox
b_pfi_path = download("https://www.dropbox.com/s/pp5a4abd6u16xy7/b_pfi_results_data.csv?dl=0","b_pfi_results_data.csv", replace=True)
ibm_pfi_path = download("https://www.dropbox.com/s/0gzicyoomzu8co0/ibm_pfi_results_data.csv?dl=0","ibm_pfi_results_data.csv", replace=True)
hra_pfi_path = download("https://www.dropbox.com/s/un7zwpov65mik2m/hra_pfi_results_data.csv?dl=0","hra_pfi_results_data.csv", replace=True)

#Load data into pandas dataframe
b_pfi_dt = pd.read_csv("b_pfi_results_data.csv", sep=";")
ibm_pfi_dt = pd.read_csv("ibm_pfi_results_data.csv", sep=";")
hra_pfi_dt = pd.read_csv("hra_pfi_results_data.csv", sep=";")
```

```python
#Analysis source: https://github.com/JosephMagiya/Kendall-Rank-Correlation-Python/blob/master/Kendall%20Rank%20Correlation.ipynb
```

```python
# Data Visualisation Pre-settings 
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sns.set_style('whitegrid')
```

```python
#Kendall tau calculation on Babushkin dataset
corr_b_pfi = b_pfi_dt.corr(method='kendall')

sns.heatmap(corr_b_pfi, xticklabels=corr_b_pfi.columns.values, yticklabels=corr_b_pfi.columns.values, cmap="viridis", annot=True).set(title='Babushkin PFI Kendall\'s tau results')
```

```python
#Kendall tau calculation on IBM dataset
corr_ibm_pfi = ibm_pfi_dt.corr(method='kendall')

sns.heatmap(corr_ibm_pfi, xticklabels=corr_ibm_pfi.columns.values, yticklabels=corr_ibm_pfi.columns.values, cmap="viridis", annot=True).set(title='IBM HRA PFI Kendall\'s tau results')
```

```python
#Kendall tau calculation on HRA dataset
corr_hra_pfi = hra_pfi_dt.corr(method='kendall')

sns.heatmap(corr_hra_pfi, xticklabels=corr_hra_pfi.columns.values, yticklabels=corr_hra_pfi.columns.values, cmap="viridis", annot=True).set(title='HRA PFI Kendall\'s tau results')
```
